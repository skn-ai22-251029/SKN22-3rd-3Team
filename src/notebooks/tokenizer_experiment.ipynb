```python
import os
import sys
from kiwipiepy import Kiwi

# 1. í”„ë¡œì íŠ¸ ë£¨íŠ¸ ê²½ë¡œ ì„¤ì • ë¡œì§
# ë…¸íŠ¸ë¶ì´ src/notebooks/ í•˜ìœ„ì— ìˆë‹¤ê³  ê°€ì •í•˜ê³  ë£¨íŠ¸ë¥¼ ì°¾ìŠµë‹ˆë‹¤.
current_path = os.getcwd()
project_root = current_path

# 'data' í´ë”ê°€ ë³´ì¼ ë•Œê¹Œì§€ ìƒìœ„ ë””ë ‰í† ë¦¬ë¡œ ì´ë™
while not os.path.exists(os.path.join(project_root, 'data')) and project_root != '/':
    project_root = os.path.dirname(project_root)

if project_root == '/':
    # Fallback: src/notebooks ì— ìˆë‹¤ê³  ê°€ì •í•˜ê³  ë‘ ë‹¨ê³„ ìƒìœ„ë¡œ
    project_root = os.path.abspath(os.path.join(os.getcwd(), "../../"))
    
sys.path.append(project_root)
print(f"ğŸ“‚ Project Root: {project_root}")

DICT_PATH = os.path.join(project_root, "src/core/tokenizer/domain_dictionary.txt")
if os.path.exists(DICT_PATH):
    print(f"âœ… ë„ë©”ì¸ ì‚¬ì „ì„ ì°¾ì•˜ìŠµë‹ˆë‹¤: {DICT_PATH}")
else:
    print(f"âŒ ë„ë©”ì¸ ì‚¬ì „ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {DICT_PATH}")
```

## í† í¬ë‚˜ì´ì € ì´ˆê¸°í™”
ë‘ ê°€ì§€ Kiwi ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•©ë‹ˆë‹¤:

```python
# ê¸°ë³¸ Kiwi ì´ˆê¸°í™”
kiwi_default = Kiwi()

# ì»¤ìŠ¤í…€ Kiwi ì´ˆê¸°í™”
kiwi_custom = Kiwi()
try:
    kiwi_custom.load_user_dictionary(DICT_PATH)
    print("âœ¨ ì»¤ìŠ¤í…€ ì‚¬ìš©ì ì •ì˜ ì‚¬ì „ ë¡œë“œ ì™„ë£Œ")
except Exception as e:
    print(f"âš ï¸ ì‚¬ì „ ë¡œë“œ ì‹¤íŒ¨: {e}")
```

## ë¹„êµ í•¨ìˆ˜

```python
def compare_tokens(text):
    tokens_def = [t.form for t in kiwi_default.tokenize(text)]
    tokens_cus = [t.form for t in kiwi_custom.tokenize(text)]
    
    print(f"ğŸ“ ì…ë ¥ ë¬¸ì¥: {text}")
    print(f"   âŒ Default: {tokens_def}")
    print(f"   âœ… Custom : {tokens_cus}")
    
    # ê²°ê³¼ê°€ ë‹¤ë¥¼ ê²½ìš° ê°•ì¡° í‘œì‹œ
    if tokens_def != tokens_cus:
        print("   ğŸ’¡ ì°¨ì´ê°€ ë°œê²¬ë˜ì—ˆìŠµë‹ˆë‹¤!")
    else:
        print("   (ê²°ê³¼ê°€ ë™ì¼í•©ë‹ˆë‹¤)")
    print("-" * 50)
```

```python
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤
examples = [
    "ì§‘ì‚¬ëŠ” ë§›ë™ì‚°ì„ ìºê¸° ìœ„í•´ ë²¤í† ë‚˜ì´íŠ¸ ëª¨ë˜ë¥¼ ìƒ€ë‹¤.",
    "ë©”ì¸ì¿¤ê³¼ ì•„ë¹„ì‹œë‹ˆì•ˆì€ ì„±ê²©ì´ ë‹¤ë¥´ë‹¤.",
    "ê³ ì–‘ì´ ì·Œì¥ì—¼ì€ ì¬ë°œì´ ì¦ë‹¤.",
    "ì¸„ë¥´ë¥¼ ë¨¹ëŠ” ë¨¼ì¹˜í‚¨ ê³ ì–‘ì´",
    "ëŸ¬ì‹œì•ˆë¸”ë£¨ëŠ” ì¡°ìš©í•œ í¸ì´ë‹¤."
]

for ex in examples:
    compare_tokens(ex)
```
