# Vector Retriever 테스트 (Semantic Search)

이 노트북은 `VectorRetriever`의 의미론적 검색, 미등록 품종에 대한 유사도 동작 등을 검증합니다.

```python
# 환경 설정
import sys
import os
from dotenv import load_dotenv

# 프로젝트 루트 경로 설정
project_root = os.path.abspath(os.path.join(os.getcwd(), "../.."))
sys.path.append(project_root)

load_dotenv(os.path.join(project_root, ".env"))
print(f"Project Root: {project_root}")
```

```python
from src.retrieval.vector_retriever import VectorRetriever

vector = VectorRetriever(version="v3", collection_name="care_guides")
print("✅ VectorRetriever initialized.")
```

## 1. 의미론적 검색 (Semantic Search)
키워드가 정확히 일치하지 않아도 의미적으로 유사한 문서를 찾는지 확인합니다.

```python
query = "활동적이고 개냥이 같은 친구 추천해줘"
results = await vector.search(query)

print(f"Query: {query}")
for i, r in enumerate(results):
    print(f"{i+1}. {r.get('name_ko')} (Score type: {r.get('score_type')})")
    print(f"   - Summary: {r.get('summary')[:50]}...")
```

## 2. 미등록 품종 (Unknown Breed)
DB에 없는 '갤럭시 캣'을 검색했을 때, Vector Search는 어떤 결과를 가져오는지 확인합니다.

```python
query = "갤럭시 캣"
results = await vector.search(query)

print(f"Query: {query}")
for i, r in enumerate(results):
    print(f"{i+1}. {r.get('name_ko')} (Score type: {r.get('score_type')})")
```
